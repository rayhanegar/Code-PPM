Terdapat beberapa cara untuk mengatasi permasalahan missing value pada data. Salah satu cara yang sederhana adalah dengan mengganti nilai NaN pada suatu fitur dengan rata-rata nilai fitur tersebut pada data lain yang bernilai bukan Nan. Perhatikan bahwa Anda harus mengganti nilai NaN dengan rata-rata data lain yang memiliki kategori (species) yang sama.

Fungsi-fungsi pada Pandas yang dapat Anda manfaatkan:


1.   **transform** untuk mengaplikasikan fungsi tertentu pada dataframe, pada permasalahan ini fungsi yang digunakan adalah fungsi **mean**
2.   **groupby** untuk mengelompokkan dataframe berdasarkan nilai kolom tertentu, pada permasalahan ini kolom yang digunakan adalah **species**
3. **fillna** untuk mengganti nilai NaN dengan nilai yang telah ditentukan




Buatlah fungsi bernama **imputasi** yang melakukan tahapan berikut:


1.   Menghitung rata-rata masing-masing kolom berdasarkan kelasnya
2.   Mengisi kolom yang berisi NaN dengan rata-rata kelas yang sesuai




```
def impute_avg(df):
  df_copy = df.copy()
  cols = df.columns
  target = cols[-1]

  # asumsi kolom terakhir adalah kolom target
  cols = cols[:-1]
  for col in cols:
    print("{col} missing values before: {nan}".format(col=col, nan=df_copy[col].isna().sum()))
    df_copy[col] = df_copy[col].fillna(df_copy.groupby(target)[col].transform('mean'))
    print("{col} missing values after: {nan}".format(col=col, nan=df_copy[col].isna().sum()))
    print()
  return df_copy
```

Buatlah sebuah dataframe baru bernama **data_imputasi** yang berisi dataset dengan nilai NaN yang sudah diganti dengan cara memanggil fungsi **imputasi**.

```
iris_df_impute = impute_avg(iris_df)
```

### 3) Normalisasi MinMax
Normalisasi bertujuan menyamakan rentang nilai pada setiap fitur. Beberapa metode pembelajaran mesin memiliki kinerja yang buruk apabila rentang nilai tiap variabel berbeda jauh.

Sebelum melakukan normalisasi, cek terlebih dahulu rentang (nilai max - nilai min) pada masing masing fitur

```
def check_minmax(df, brief=False):
  df_copy = df.copy()
  cols = df.columns[:-1]
  range_arr = []
  for col in cols:
    max = df_copy[col].max()
    min = df_copy[col].min()
    range_arr.append(max-min)
    if(~brief):
      print("{col} range is: {range}".format(col=col, range=range_arr[-1]))
  return range_arr
```

```
iris_range = check_minmax(iris_df)
```

Terlihat bahwa masing-masing fitur memiliki rentang yang berbeda, meskipun tidak terlalu signifikan. Pada praktikum ini Anda akan mengimplementasikan normalisasi MinMax untuk menyamakan rentang setiap fitur menjadi satu (1). Persamaan dari fungsi normalisasi MinMax adalah sebagai berikut:

$x'=\frac{x-min(x)}{max(x)-min(x)}$

dimana $x'$ merupakan nilai ternormalisasi, $x$ adalah nilai asal, $min(x)$ merupakan nilai minimum pada suatu fitur, dan $max(x)$ merupakan nilai maksimum pada suatu fitur.

Implementasikan metode normalisasi MinMax pada fungsi bernama **minmax**.


```
def minmax(df):
  df_copy = df.copy()
  cols = df.columns[:-1]
  for col in cols:
    max = df_copy[col].max()
    min = df_copy[col].min()

    df_copy[col] = (df_copy[col] - min) / (max - min)
  return df_copy
```

3. Jelaskan perbedaan hasil normalisasi **MinMax** dan **ZScore**. Petunjuk : cek nilai rentang, rata-rata dan standar deviasi dari **data_zscore**.


```
def zscore(df):
  #tuliskan jawaban nomor 1 di sini
  df_copy = df.copy()
  cols = df_copy.columns[:-1]
  for col in cols:
    mean = np.mean(df_copy[col])
    stdev = np.std(df_copy[col])
    df_copy[col] = (df_copy[col] - mean) / stdev
  return df_copy
```

### Method zscore(df)
Method ini merupakan implementasi dari bentuk normalisasi Z-Score (Standard Normalization) dengan formula $ x' = \frac{x - \mu}{σ}$. Dengan demikian, data yang sudah ditransformasi akan memiliki nilai mean = 0 dan nilai standar deviasi = 1.

- df_copy = df.copy() digunakan untuk membuat DataFrame yang akan dimanipulasi, menjaga nilai asli DataFrame awal.
- cols = df_copy.columns[:-1] digunakan untuk mengambil kolom/fitur DataFrame kecuali kolom terakhir.
- for col in cols digunakan untuk melakukan iterasi untuk setiap kolom/fitur.
- np.mean() digunakan untuk mendapatkan nilai mean dari Series yang dijadikan argumen.
- np.std() digunakna untuk mendapatkan nilai standar deviasi dari Series yang dijadikan argumen.
- df_copy[col] = (df_copy[col] - mean)/stdev digunakan untuk melakukan proses normalisasi untuk masing-masing fitur DataFrame.

Fungsi ini akan mengembalikan DataFrame yang sudah melalui normalisasi Z-Score.


```
data_zscore = zscore(iris_df_impute)
```

### Komparasi Normalisasi MinMax dan Z-Score

#### Range
- Normalisasi MinMax menghasilkan nilai range konsisten maximum 1.0 dikarenakan nilai berada pada rentang 0.0--1.0
- Normalisasi Z-Score menghasilkan nilai range yang bervariasi, dan tidak terbatas pada 0.0--1.0. Hal ini dikarenakan normalisasi Z-Score berusaha untuk menjaga/representatif terhadap jarak data awal.

#### Mean
- Normalisasi MinMax menghasilkan nilai mean yang bervariasi sesuai dengan kondisi data.
- Normalisasi Z-Score menghasilkan nilai mean yang sangat kecil dan mendekati 0.

#### Standar Deviasi
- Normalisasi MinMax menghasilkan nilai standar deviasi yang bervariasi.
- Normalisasi Z-Score menghasilkan nilai standar deviasi yang konsisten untuk semua fitur, yaitu 1.
