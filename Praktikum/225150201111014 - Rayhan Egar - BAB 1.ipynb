### 4) Membaca format CSV

File CSV dapat diuraikan (parsing) ke dalam bentuk list, dictionary, atau dataframe.

Tuliskan code pada cell di bawah ini untuk melakukan parsing csv ke dalam bentuk list

```
def csv_list(filename):
  data = []
  with open('kontak.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
      data.append(row)
    data.pop(0)
    return data
```

Cek apakah fungsi csv_list dapat berjalan dengan baik

```
list_kontak = csv_list('kontak.csv')
print(list_kontak)
```

Tuliskan code pada cell di bawah ini untuk melakukan parsing csv ke dalam bentuk dictionary

```
def csv_dict(filename):
  data = []
  with open(filename) as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
      data.append(row)
  return data
```

Cek apakah fungsi csv_dict dapat berjalan dengan baik




```
kontak_dict = csv_dict('kontak.csv')
print(kontak_dict)
```

## 4. Hitung rata-rata nilai per variabel untuk setiap data
Perhitungan nilai rata-rata untuk setiap feature pada masing-masing data dapat dilakukan dengan bantuan fungsi 'np.mean()' pada NumPy dengan argument berisikan kolom dari variabel/feature yang diinginkan.

Langkah ini memanfaatkan atribut '.columns' dari kelas DataFrame dan melakukan iterasi untuk setiap variabel menggunakan for-loop dengan langkah-langkah sebagai berikut.
1. 'variables = list(glass_df.columns)' digunakan untuk mendapatkan nama-nama kolom dari variable glass_df dan menyimpannya dalam bentuk list pada variabel bernama 'variables'.
2. 'variables = variables[:-1]' digunakan untuk menghapus entri terakhir dari variables. Hal ini dikarenakan kolom terakhir ('Type') merupakan data kualitatif dan tidak bisa dicari rata-ratanya.
3. 'for variable in variables' digunakan untuk melakukan iterasi dan menjalankan loop body untuk setiap elemen di dalam list 'variables'.
4. Perhitungan rata-rata menggunakan function dari NumPy, yaitu '.mean()' dengan memasukkan 'glass_df[variable]' sebagai argumen. Hasil pemanggilan 'np.mean()' ditampilkan dalam bentuk formatted string.

Alternatif lainnya dengan memanggil .mean() langsung pada obyek DataFrame

```
variables = list(glass_df.columns)
variables = variables[:-1]

for variable in variables:
  print("Rata-rata variabel {variable} = {average}".format(variable=variable,average=np.mean(glass_df[variable])))

# Alternatif
print()
print(glass_df.mean())
```

## 5. Hitung rata-rata nilai per variabel untuk setiap data, dikelompokkan berdasarkan 'Type'

```
avgs = glass_df.groupby('Type').mean()
print(avgs)
```
