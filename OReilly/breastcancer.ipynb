```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
```

```python
cancer = load_breast_cancer()
cancerdf = pd.DataFrame(cancer.data, columns=cancer.feature_names)
cancerdf["target"] = cancer.target
```

```python
cancerdf.head()
```

```python
cancer_0 = cancerdf[cancerdf['target']==0]
cancer_1 = cancerdf[cancerdf['target']==1]
sampled_cancer_0 = resample(cancer_0, replace=False, n_samples=25)
sampled_cancer_1 = resample(cancer_1, replace=False, n_samples=25)
cancer_50 = pd.concat([sampled_cancer_0, sampled_cancer_1])
```
